```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Завантаження даних
data = pd.read_csv('TransnetBW.csv', index_col='Date', parse_dates=True)

# Перетворення даних у вектори (всі інтервали як окремі фічі)
data_flattened = data.values.reshape(-1, 96)  # 96 часових інтервалів

# Нормалізація даних
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data_flattened)

# Розділення на навчальні та тестові набори
X_train, X_test, y_train, y_test = train_test_split(
    data_scaled[:, :-1],  # Всі часові інтервали крім останнього
    data_scaled[:, -1],   # Останній часовий інтервал як цільова змінна
    test_size=0.2, random_state=42
)
print(data)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten

# Створення моделі
cnn_model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
    Flatten(),
    Dense(50, activation='relu'),
    Dense(1)
])

# Компіляція моделі
cnn_model.compile(optimizer='adam', loss='mse')

# Навчання моделі
cnn_model.fit(X_train, y_train, epochs=10, verbose=1, validation_split=0.2)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, MultiHeadAttention, Dropout, LayerNormalization, Dense

# Визначення спрощеної трансформерної моделі
def transformer_model(input_shape):
    inputs = Input(shape=input_shape)
    # Тут можемо використовувати MultiHeadAttention
    x, _ = MultiHeadAttention(num_heads=2, key_dim=2)(inputs, inputs, return_attention_scores=True)
    x = Dropout(0.1)(x)
    x = LayerNormalization(epsilon=1e-6)(x)
    x = Dense(50, activation='relu')(x)
    outputs = Dense(1)(x)
    model = Model(inputs=inputs, outputs=outputs)
    return model

# Створення моделі
input_shape = (X_train.shape[1], 1)
trans_model = transformer_model(input_shape)
trans_model.compile(optimizer='adam', loss='mse')

# Навчання моделі
trans_model.fit(X_train, y_train, epochs=10, verbose=1, validation_split=0.2)
```
