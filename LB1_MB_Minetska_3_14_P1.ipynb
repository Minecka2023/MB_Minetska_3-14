2. Постройте график функциий `f(x) = x^2` и `f(x) = 10*sin(x)`, а также их суммы для x от -5 до 12. Не забудьте обозначить оси и легенду.

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt

# Функции
def f1(x):
    return x**2

def f2(x):
    return 10 * np.sin(x)

# Генерация значений x от -5 до 12 с небольшим шагом
x_values = np.linspace(-5, 12, 1000)

# Вычисление значений функций для каждого x
y1_values = f1(x_values)
y2_values = f2(x_values)
y_sum = y1_values + y2_values

# Построение графиков
plt.figure(figsize=(10, 6))

plt.plot(x_values, y1_values, label='f(x) = х²', color='blue')
plt.plot(x_values, y2_values, label='f(x) = 10*sin(x)', color='orange')
plt.plot(x_values, y_sum, label='10*sin(x)+х²', color='green')

# Подписи и легенда
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Отображение графика
plt.grid(True)
plt.show()
```

3. Постройте верхнюю половину сферы в 3d с центром в точке $(0, 0, 0)$ и радиусом $R = 1$.

**Подсказка** Сфера однозначно задается функцией вида $x^2 + y^2 + z^2 = R^2$

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Создание данных для верхней половины сферы
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi / 2, 100)

theta, phi = np.meshgrid(theta, phi)

# Параметры сферы в сферических координатах
R = 1
r = R * np.sin(phi)
x = r * np.cos(theta)
y = r * np.sin(theta)
z = R * np.cos(phi)

# Создание и настройка 3D графика
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='binary', alpha=0.7)

# Настройка осей и вида
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 40)
fig
# Отображение графика
plt.show()
```

4. Напишите код, чтобы нарисовать график, приведенный ниже.

```python
import numpy as np
import matplotlib.pyplot as plt

# Создание данных для полукруга
theta = np.linspace(0, np.pi, 1000)
x_circle = 10 * np.cos(theta)
y_circle = -10 * np.sin(theta)  # Отрицательное значение для части полукруга внизу

# Создание данных для вертикальной линии
x_line = np.full(100, -10)  # Вертикальная линия на отметке -10 по оси X
y_line = np.linspace(0, 10, 100)

# Создание данных для второй вертикальной линии из точек
x_dotted_line = np.full(11, 10)  # Вертикальная линия на отметке 10 по оси X из 11 точек
y_dotted_line = np.linspace(0, 10, 11)

# Создание данных для горизонтальной линии
x_horizontal_line = np.linspace(-10, 10, 100)
y_horizontal_line = np.full(100, 10)  # Горизонтальная линия на отметке 10 по оси Y

# Создание и настройка графика
plt.figure(figsize=(8, 8))

# Отображение полукруга
plt.plot(x_circle, y_circle, label='Полукруг', color='red')

# Отображение первой вертикальной линии
plt.plot(x_line, y_line, linestyle='--', color='darkorange', linewidth=2, label='Вертикальная линия 1')

# Отображение второй вертикальной линии из точек
plt.plot(x_dotted_line, y_dotted_line, marker='o', linestyle='', color='green', markersize=4, label='Вертикальная линия 2')

# Отображение горизонтальной линии
plt.plot(x_horizontal_line, y_horizontal_line, linestyle='-', color='blue', linewidth=2, label='Горизонтальная линия')

# Настройка осей
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(False)

# Отображение графика
plt.show()
```
